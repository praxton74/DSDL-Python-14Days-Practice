```python
plt.bar(a,b,label="KIET",color="RED")
# plt.bar(a,b)
plt.xlabel("Subject")
plt.ylabel("Marks")
plt.legend()
```

```python
p=[10,56,55,41,22]
q=[17,16,30,9,33]
r=[27,36,21,4,95]
```
